```
data1=pd.read_csv('/home/jovyan/work/churn_prediction.csv')
```

```
data1.isnull().sum()
```

```
data1["gender"].fillna( method ='ffill', inplace = True)
```

```
data1["dependents"].fillna( method ='ffill', inplace = True)
```

```
data1["occupation"].fillna( method ='ffill', inplace = True)
```

```
data1["city"].fillna( method ='ffill', inplace = True)
```

```
data1["days_since_last_transaction"].fillna( method ='ffill', inplace = True)
```

```
data1.head(5)
```

```

from sklearn.preprocessing import OrdinalEncoder
```

```
ord_enc = OrdinalEncoder()
```

```
data1['gender']=ord_enc.fit_transform(data1[['gender']])
data1['occupation']=ord_enc.fit_transform(data1[['occupation']])
```

```
x=data1[['vintage','age','gender','dependents','occupation','city','customer_nw_category','branch_code','days_since_last_transaction','current_balance','previous_month_end_balance','average_monthly_balance_prevQ','average_monthly_balance_prevQ2','current_month_credit','previous_month_credit','current_month_debit','previous_month_debit','current_month_balance','previous_month_balance']]
y=data1['churn']
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 42)
```

```
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```
y_pred = classifier.predict(X_test)
```

```
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
